```python
import seaborn as sns
import pandas  as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
df = data.groupby(['day', 'time'],   observed=False).size().unstack().plot(kind='bar')
print(df)
plt.xlabel('sex')
plt.ylabel('count')
plt.title('Total sizes by Day')
plt.show()

df=data.groupby('day',observed=False)['total_bill'].sum().reset_index()
print(df)
plt.bar(df['day'],df['total_bill'])
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()

df=data.groupby('day',observed=False)['size'].sum().reset_index()
print(df)
plt.bar(df['day'],df['size'])
plt.xlabel('Day')
plt.ylabel('Size')
plt.title('Total sizes by Day')
plt.show()

df = data.groupby(['day', 'sex'],  observed=False).size().unstack().plot(kind='bar')
print(df)
plt.xlabel('day')
plt.ylabel('sex')
plt.title('Total visited by Day')
plt.show()

df = data.groupby(['day', 'time'],  observed=False).size().unstack().plot(kind='bar')
print(df)
plt.xlabel('sex')
plt.ylabel('count')
plt.title('Total sizes by Day')
plt.show()

df = data.groupby(['sex', 'smoker'],  observed=False).size().nlargest().plot(kind='bar')
print(df)
plt.xlabel('sex')
plt.ylabel('count')
plt.title('smoker and non smokers')
plt.show()

filter_by=data[(data['sex']=='Male') | (data['sex']=='Female')]
print(filter_by)
df = filter_by.groupby(['sex', 'smoker'],  observed=False).size().unstack().plot(kind='bar')
plt.xlabel('sex')
plt.ylabel('count')
plt.title('smoker and non smokers')
plt.show()
```
